```python
from __future__ import division
import pylab as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
import seaborn as sns
import os
```

# Ecoli datasets

```python
data = pd.read_csv('../summary_data/ecoli_summary.csv')
print(data.columns)
data["Trim Effeciency"] = data["Base - TP TRIM"]/(data["Base - TP TRIM"] + data["Base - FP TRIM"])
data['total_corrections']=data['Base - TP']+ data['Base - FP']
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tool='bfc'
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Coverage", y="Base Gain", kind="point",data = data[(data['Wrapper Name'] == tool)]
, hue='Kmer Size')
g.set(xlabel='Coverage', ylabel='Gain', title=tool)
g=sns.despine()
#plt.show(g)
plt.savefig("../figures/ecoli_gain_tool_"+tool+".png", bbox_inches='tight')
```
