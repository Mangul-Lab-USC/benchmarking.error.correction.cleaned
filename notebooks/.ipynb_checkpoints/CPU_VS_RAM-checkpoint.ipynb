```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import rc
import numpy as np

df = pd.read_csv("../summary_data/WGS_CPU_VS_RAM.csv")
df['RAM_megabytes'] = df["RAM_megabytes"] / 1000
df = df.rename(columns={"RAM_megabytes": "RAM_Gigabytes"})
df["Hours"] = df["CPU"]/360


df_20= df.loc[df["kmer_size"] == 20]
df_22= df.loc[df["kmer_size"] == 22]
df_24= df.loc[df["kmer_size"] == 24]
df_26= df.loc[df["kmer_size"] == 26]
df_28= df.loc[df["kmer_size"] == 28]
df_30= df.loc[df["kmer_size"] == 30]


print("Kmer size 20 avg RAM (gigabytes) across tools", df_20.sum()["RAM_Gigabytes"] / df_20.count()["Tool"])
print("Kmer size 22 avg RAM (gigabytes) across tools",df_22.sum()["RAM_Gigabytes"] / df_22.count()["Tool"])
print("Kmer size 24 avg RAM (gigabytes) across tools",df_24.sum()["RAM_Gigabytes"] / df_24.count()["Tool"])
print("Kmer size 26 avg RAM (gigabytes) across tools",df_26.sum()["RAM_Gigabytes"] / df_26.count()["Tool"])
print("Kmer size 28 avg RAM (gigabytes) across tools",df_28.sum()["RAM_Gigabytes"] / df_28.count()["Tool"])
print("Kmer size 30 avg RAM (gigabytes) across tools",df_30.sum()["RAM_Gigabytes"] / df_30.count()["Tool"])

print("avg RAM (gigabytes) across tools for all kmer sizes",df.sum()["RAM_Gigabytes"] / df.count()["Tool"])

print("\n\n")


print("Kmer size 20 avg CPU across tools", df_20.sum()["CPU"] / df_20.count()["Tool"])
print("Kmer size 22 avg CPU across tools",df_22.sum()["CPU"] / df_22.count()["Tool"])
print("Kmer size 24 avg CPU across tools",df_24.sum()["CPU"] / df_24.count()["Tool"])
print("Kmer size 26 avg CPU across tools",df_26.sum()["CPU"] / df_26.count()["Tool"])
print("Kmer size 28 avg CPU across tools",df_28.sum()["CPU"] / df_28.count()["Tool"])
print("Kmer size 30 avg CPU across tools",df_30.sum()["CPU"] / df_30.count()["Tool"])

print("avg CPU across tools for all kmer sizes",df.sum()["CPU"] / df.count()["Tool"])

df.head()
```

```python
print("Kmer size 20 tool with longest CPU time:", df_20.max()["CPU"]," Tool:", df_20.max()["Tool"])
print("Kmer size 20 tool with shortest CPU time:", df_20.min()["CPU"]," Tool:", df_20.min()["Tool"] )

print("Kmer size 22 tool with longest CPU time:", df_22.max()["CPU"]," Tool:", df_22.max()["Tool"])
print("Kmer size 22 tool with shortest CPU time:", df_22.min()["CPU"]," Tool:", df_22.min()["Tool"] )

print("Kmer size 24 tool with longest CPU time:", df_24.max()["CPU"]," Tool:", df_24.max()["Tool"])
print("Kmer size 24 tool with shortest CPU time:", df_24.min()["CPU"]," Tool:", df_24.min()["Tool"] )

print("Kmer size 26 tool with longest CPU time:", df_26.max()["CPU"]," Tool:", df_26.max()["Tool"])
print("Kmer size 26 tool with shortest CPU time:", df_26.min()["CPU"]," Tool:", df_26.min()["Tool"] )

print("Kmer size 28 tool with longest CPU time:", df_28.max()["CPU"]," Tool:", df_28.max()["Tool"])
print("Kmer size 28 tool with shortest CPU time:", df_28.min()["CPU"]," Tool:", df_28.min()["Tool"] )

print("Kmer size 30 tool with longest CPU time:", df_30.max()["CPU"]," Tool:", df_30.max()["Tool"])
print("Kmer size 30 tool with shortest CPU time:", df_30.min()["CPU"]," Tool:", df_30.min()["Tool"] )

print("\n")

print("tool with longest CPU time of all kmers:", df.max()["CPU"]," Tool:", df.max()["Tool"])
print("tool with shortest CPU time of all kmers:", df.min()["CPU"]," Tool:", df.min()["Tool"] )



```

```python
color_dict=dict({'bfc':'purple','bless':'orange','coral':'brown','fiona':'gray','lighter':'pink','musket':'blue','pollux':'yellow','racer':'green','reckoner':'red','sga':'black'})
print (color_dict)
```

```python
ax = sns.barplot(x="kmer_size", y="RAM_Gigabytes", hue="Tool", palette=color_dict, data=df)
g=sns.set_style("white")
g=sns.set_context("poster")

# Custom X axis
plt.xlabel("Kmer size", fontsize=18, labelpad=12)
 
# Custom Y axis
plt.ylabel("Maxmimum RAM (gigabytes)",fontsize=20, labelpad=12)

#  legend
plt.legend(bbox_to_anchor=(1.6, 1),frameon=False)

g=sns.despine()
plt.savefig("../figures/All_kmer_Ram.pdf", bbox_inches='tight')
```
