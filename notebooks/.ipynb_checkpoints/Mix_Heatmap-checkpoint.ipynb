```python
from __future__ import division
import pylab as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
import seaborn as sns
import os
```

```python
data = pd.read_csv('../summary_data/mix_summary.csv')
print(data.columns)
data["Trim Effeciency"] = data["Base - TP TRIM"]/(data["Base - TP TRIM"] + data["Base - FP TRIM"])
data['total_corrections']=data['Base - TP']+ data['Base - FP']
data.head()
```

# CHANGE KMER VALUE HERE 

```python
KMER=30

data = data[(data['Kmer Size'] == KMER)]
```

## Total number of corrections 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
g=sns.set_style("white")
g=sns.set_context("talk")

g=sns.catplot(x="total_corrections", y="Tool" , col = "Error", col_wrap = 4,data = data, kind = "bar")
g.set(xlabel='Total number of corrections', ylabel='Methods', xscale='log')
g.fig.tight_layout()
g=sns.despine()
plt.show(g)


plt.savefig("../figures/mix_total_corrections.pdf", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Error", y="Base Gain", kind="point",data=data, hue='Tool')
, hue='Tool')
g.set(xlabel='Error', ylabel='Base Gain')
g.set_xticklabels(rotation=36)
g=sns.despine()
#plt.show(g)
plt.savefig("../figures/mix_Gain_vs_error.png", bbox_inches='tight')
```

```python
result = data.pivot(index='Tool', columns='Error', values='Base Gain')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.heatmap(result, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
plt.savefig("../figures/mix_heatmap_gain_vs_error.png", bbox_inches='tight')
```
