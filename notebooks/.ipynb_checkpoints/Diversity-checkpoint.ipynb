```python
from __future__ import division
import pylab as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
import seaborn as sns
import os
```

# Mix datasets

```python
data = pd.read_csv('../summary_data/diversity_summary.csv')
print(data.columns)
data["Trim Effeciency"] = data["Base - TP TRIM"]/(data["Base - TP TRIM"] + data["Base - FP TRIM"])
data['total_corrections']=data['Base - TP']+ data['Base - FP']
#data = data[(data['Kmer Size'] >19)]
#data = data[(data['Kmer Size'] <= 30)]
data
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="dist", y="Base Gain", kind="point",data = data , hue='Tool')

tick = ticker.ScalarFormatter(useOffset=False, useMathText=True)
tick.set_powerlimits((0,0))
tc = [u"${}$".format(tick.format_data(x)) for x in data['dist']]
g.set(xlabel='Dist', ylabel='Gain',xticklabels=tc)

g.set_xticklabels(rotation=36)
g=sns.despine()
#plt.show(g)
plt.savefig("../figures/diverse_gain.png", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
g=sns.set_style("white")
g=sns.set_context("talk")

g=sns.catplot(x="total_corrections", y="Tool" , col = "dist", col_wrap = 3,data = data, kind = "bar")
g.set(xlabel='Total number of corrections', ylabel='Methods', xscale='log')
g.fig.tight_layout()
g=sns.despine()
plt.show(g)


plt.savefig("../figures/diverse_total_corrections.pdf", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")

result = data.pivot(index='Tool', columns='dist', values='Base Gain')

tick = ticker.ScalarFormatter(useOffset=False, useMathText=True)
tick.set_powerlimits((0,0))
tc = [u"${}$".format(tick.format_data(x)) for x in data['dist'].unique()]

g=sns.heatmap(result, fmt="g", cmap='gist_ncar',center=0,linewidths=.5,xticklabels=tc)
g.set(title='Base Gain')

plt.savefig("../figures/diverse_heatmap_gain_vs_error.png", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")

result_PPV = data.pivot(index='Tool', columns='dist', values='Base Precision')


tick = ticker.ScalarFormatter(useOffset=False, useMathText=True)
tick.set_powerlimits((0,0))
tc = [u"${}$".format(tick.format_data(x)) for x in data['dist'].unique()]


g=sns.heatmap(result_PPV, fmt="g", cmap='gist_ncar',center=0,linewidths=.5, xticklabels=tc)
g.set(title='Base Precision')

plt.savefig("../figures/diverse_heatmap_ppv_vs_error.png", bbox_inches='tight')
```
