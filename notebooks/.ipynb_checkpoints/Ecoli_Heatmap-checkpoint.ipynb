```python
from __future__ import division
import pylab as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
import seaborn as sns
import os
```

```python
data = pd.read_csv('../summary_data/ecoli_summary.csv')
print(data.columns)
data["Trim Effeciency"] = data["Base - TP TRIM"]/(data["Base - TP TRIM"] + data["Base - FP TRIM"])
data['total_corrections']=data['Base - TP']+ data['Base - FP']
data.head()
```

# CHANGE KMER VALUE HERE

```python
KMER=30

data = data[(data['Kmer Size'] == KMER)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")

g=sns.catplot(x="total_corrections", y="Tool", kind="bar",data= data[(data['Coverage'] == 1)]);
g.set(xlabel='Total number of corrections', ylabel='Methods', title='Kmer '+str(KMER)+' Coverage 1')
g.fig.tight_layout()
g=sns.despine()
plt.savefig("../figures/ecoli_total_corrections_vs_tool_cov1_kmer"+str(KMER)+".png", bbox_inches='tight')
```

```python
result = data.pivot(index='Tool', columns='Coverage', values='Base Gain')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.heatmap(result, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
g.set(xlabel='Coverage', ylabel='Tool',  title='Gain - Kmer '+str(KMER))
plt.savefig("../figures/ecoli_heatmap_gain_vs_cov_kmer"+str(KMER)+".png", bbox_inches='tight')
```

```python
result_PPV = data.pivot(index='Tool', columns='Coverage', values='Base Precision')

import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.heatmap(result_PPV, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
g.set(xlabel='Coverage', ylabel='Tool',  title='Base Precision - Kmer '+str(KMER))

plt.savefig("../figures/ecoli_heatmap_ppv_vs_cov_kmer"+str(KMER)+".png", bbox_inches='tight')
```
